```python
drift_rate = ((spy['Close'].pct_change().sum())/(spy['Close'].pct_change().count())).iloc[-1
                                                                                          ]
drift_rate
```

```python
volatility = (spy['Close'].pct_change().std()).iloc[-1]
vol = volatility * np.sqrt(252)
```

```python
delta_T = (float)(1/252)
```

```python
Num_simulations= 10000
```

```python
N_steps = 6 # Number of Market Days between current Date and Maturity
```

### **Barriers Defined**

```python
up_rate=6 # 6% more than current price
barrier_up = current_price*(1+float(up_rate/100))
barrier_up
```

```python
down_rate=6 # 6% more than current price
barrier_down = current_price*(1-float(down_rate/100))
barrier_down
```

# **Geometric Brownian Motion**

image.png

```python
Maturity_price=[]
count=0
for i in range(0,Num_simulations):
  count+=1
  print(count)
  S = current_price
  path=[S]
  for step in range(0,N_steps):
    S = S * np.exp((drift_rate - (1/2)*vol*vol)*delta_T + vol*np.sqrt(delta_T)*np.random.normal(0, 1))
    path.append(S)
  Maturity_price.append(path)
# Maturity_price
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
for i in range(Num_simulations):
    plt.plot(Maturity_price[i], linewidth=1)


plt.axhline(y=barrier_up, color='red', linestyle='--', linewidth=2, label='Upper Barrier')
plt.axhline(y=barrier_down, color='blue', linestyle='--', linewidth=2, label='Lower Barrier')

plt.title("Monte Carlo Simulations of Stock Price")
plt.xlabel("Days")
plt.ylabel("Simulated Price")
plt.grid(True)
plt.show()
```

## **Call Option:**

```python
strike_price = calls[calls['strike'] == int(current_price)+1]['strike'].iloc[-1]
strike_price
```

```python
calls[calls['strike'] == int(current_price)+1]['lastPrice']
```

```python
barrier_type = input("Enter barrier type (a-f): ").strip().lower()
```

```python

payoff = []

for p in Maturity_price:
    final_price = p[-1]
    crossed_up = any(price >= barrier_up for price in p)
    crossed_down = any(price <= barrier_down for price in p)

    # Barrier logic
    if barrier_type == 'a':  # Knock-In & Up
        if crossed_up:
            payoff.append(max(final_price - strike_price, 0))
        else:
            payoff.append(0)

    elif barrier_type == 'b':  # Knock-In & Down
        if crossed_down:
            payoff.append(max(final_price - strike_price, 0))
        else:
            payoff.append(0)

    elif barrier_type == 'c':  # Knock-Out & Up
        if crossed_up:
            payoff.append(0)
        else:
            payoff.append(max(final_price - strike_price, 0))

    elif barrier_type == 'd':  # Knock-Out & Down
        if crossed_down:
            payoff.append(0)
        else:
            payoff.append(max(final_price - strike_price, 0))

    elif barrier_type == 'e':  # Double Knock-In
        if crossed_up and crossed_down:
            payoff.append(max(final_price - strike_price, 0))
        else:
            payoff.append(0)

    elif barrier_type == 'f':  # Double Knock-Out
        if crossed_up or crossed_down:
            payoff.append(0)
        else:
            payoff.append(max(final_price - strike_price, 0))

# payoff = [max(float((sum(p)/(len(p))) - strike_price), 0) for p in Maturity_price]
# # payoff
```

```python
Avg_payoff = (sum(payoff))/len(payoff)
Avg_payoff
```

```python
Call_price = Avg_payoff*np.exp(-r*T)
Call_price
```

## **Conclusion**

Modeled Price = $4.8 (Num_Simulations=10000) (Knock-Out & Up)

Actual Price = $4.77

As (Modeled Price > Actual), Hence the option is underpriced(good for buyers)
